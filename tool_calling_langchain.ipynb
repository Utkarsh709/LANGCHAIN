```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

```python
#tool create
@tool
def multiply(a:int,b:int)->int:
  """ multipy two numbers """
  return a*b
```

```python
print(multiply.invoke({'a':3,'b':4}))
```

```python
multiply.name
```

```python
multiply.description
```

```python
multiply.args
```

```python
llm=ChatOpenAI()
```

```python
llm_with_tool=llm.bind_tools([multiply])
```

```python
llm_with_tool.invoke('who are you?')
```

```python
query=HumanMessage('can you multiply 4 with 16')
```

```python
messages=[query]
```

```python
messages
```

```python
result=llm_with_tool.invoke(messages)
```

```python
messages.append(result)
```

```python
messages
```

```python
tool_result=multiply.invoke(result.tool_calls[0])
```

```python
messages.append(tool_result)
```

```python
messages
```

```python
llm_with_tool.invoke(messages).content
```

```python
#tool create
@tool
def get_conversion_factor(base_currency:str,target_currency:str) ->float:
  """ This function fetches the currency conversion factor btw base currency and
  target currency"""
  API_KEY = "c754eab14ffab33112e380ca"
  url=f"https://v6.exchangerate-api.com/v6/{API_KEY}/pair/{base_currency}/{target_currency}"
  response=requests.get(url)
  return response.json()

@tool
def convert(base_currency_value:int,conversion_rate:float)->float:
  """ given a currency conversion rate thid function calculates the target currency value from
  a given base currency value """
  return base_currency_value*conversion_rate




```

```python
get_conversion_factor.invoke({'base_currency':'USD','target_currency':'INR'})
```

```python
convert.invoke({'base_currency_value':10,'conversion_rate':85.16})
```

```python
llm=ChatOpenAI()
```
